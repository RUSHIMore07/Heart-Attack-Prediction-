About data set
This database contains 76 attributes, but all published experiments refer to using a subset of 14 of them. In particular, the Cleveland database is the only one that has been used by ML researchers to
this date.The "target" field refers to the presence of heart disease in the patient. It is integer valued 0 = no/less chance of heart attack and 1 = more chance of heart attack

- Attribute Information
- 1) age
- 2) sex
- 3) chest pain type (4 values)
- 4) resting blood pressure
- 5) serum cholestoral in mg/dl
- 6)fasting blood sugar > 120 mg/dl
- 7) resting electrocardiographic results (values 0,1,2)
- 8) maximum heart rate achieved
- 9) exercise induced angina
- 10) oldpeak = ST depression induced by exercise relative to rest
- 11) the slope of the peak exercise ST segment
- 12) number of major vessels (0-3) colored by flourosopy
- 13) thal: 0 = normal; 1 = fixed defect; 2 = reversable defect
- 14) target: 0= less chance of heart attack 1= more chance of heart attack

- Dataset is taken for learning purpose. Source of the data : https://archive.ics.uci.edu/ml/datasets/Heart+Disease

The heart has several major vessels, including the aorta, vena cava, and pulmonary arteries and veins:
- Aorta: A major blood vessel that connects directly to the heart
- Superior vena cava: A major blood vessel that connects directly to the heart
- Inferior vena cava: A major blood vessel that connects directly to the heart
- Pulmonary arteries: Take oxygen-poor blood from the heart to the lungs
- Pulmonary veins: Bring oxygen-rich blood from the lungs to the heart 

These vessels are known as the great vessels, and they are an important part of the circulatory system. They send blood between the heart and lungs (pulmonary circuit) and between the heart and body (systemic circuit). The great vessels deliver deoxygenated blood from the body to the heart, which then pumps the blood into the lungs for oxygenation. The oxygenated blood is then returned to the left side of the heart and distributed back to the body.

### Distribution of ST segment slope and CAD(Slope) by Target

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='slope', hue='target', data=data, palette='Set2')
plt.title('Distribution of Slope by Target')
plt.xlabel('Slope')
plt.ylabel('Count')
plt.legend(title='Target', labels=['Less chance of heart attack', 'More chance of heart attack'])
plt.show()
```

This suggests that an increase in the slope of the peak exercise ST segment may be associated with an elevated risk of a heart attack, highlighting its potential role as a predictor of heart health and cardiovascular risk.

ST segment slope and CAD:

During exercise, the heart demands more oxygen. In people with CAD, narrowed coronary arteries may limit blood flow to the heart muscle, leading to oxygen deficiency.
This oxygen shortage can cause the ST segment to:
Downslope (slant downwards): This is considered abnormal and might indicate myocardial ischemia, a condition where the heart muscle isn't receiving enough oxygen-rich blood.
Upslope (slant upwards) or flat: These are generally considered normal responses during exercise.

As observed, individuals with thalassemia type 2 (thal = 2) have the highest count in the category indicating a higher chance of a heart attack ,

Thalassemias are a group of inherited blood disorders that primarily affect hemoglobin, the protein in red blood cells that carries oxygen. While not directly causing heart attacks, thalassemia can increase the risk of heart complications, especially heart failure, which can be a risk factor for heart attack. Here's a breakdown of the connection:

Thalassemias and Heart Risks:

Iron Overload:
The main risk factor for heart complications in thalassemia is iron overload.
Regular blood transfusions, a common treatment for thalassemia, can lead to iron buildup in the body over time.
Excess iron can damage organs, including the heart muscle, leading to heart failure.

### Machine Learning Model 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Split data into features (X) and target variable (y)
X = data.drop(columns=['target'])
y = data['target']

# Split data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Instantiate logistic regression model
logreg = LogisticRegression()

# Fit the model on the training data
logreg.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = logreg.predict(X_test)


# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate Random Forest Classifier
rf_classifier = RandomForestClassifier()

# Train the model
rf_classifier.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = rf_classifier.predict(X_test)

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import lime
import lime.lime_tabular
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression

# Train a Logistic Regression Classifier on the dataset
LogisticR = LogisticRegression(random_state=42)
LogisticR.fit(X_train, y_train)

# Instantiate LIME explainer
explainer = lime.lime_tabular.LimeTabularExplainer(X_train.values, 
                                                   feature_names=X_train.columns.tolist(),
                                                   class_names=['Less chance of heart attack', 'More chance of heart attack'],
                                                   discretize_continuous=True,
                                                   random_state=42)

instance_to_explain = np.array([[42, 1, 1, 120, 295, 0, 1, 162, 0, 0.0, 2, 0, 2]])

# Generate explanations for the instance using Logistic Regression classifier
explanation = explainer.explain_instance(instance_to_explain[0], 
                                          LogisticR.predict_proba, 
                                          num_features=13)

# Visualize explanations as text in the notebook
explanation.show_in_notebook()
```
